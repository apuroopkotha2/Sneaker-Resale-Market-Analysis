```python
# read the csv file 
df = pd.read_csv('2021_Data.csv')
df.shape
df.head()
```

```python
# grouoing the data by SKU, Size, Condition 
dff = df.groupby(['SKU', 'SIZE', 'CONDITION'])
dff.head()
```

```python
# grouped data nad assigning the value to the columns 
grouped_data = dff.agg({
    'BRAND': 'first',  
    'COLOR': 'first',  
    'SILHOUETTE': 'first',  
    'IS_COLLAB': 'first',  
    'SOLD_PRICE': 'mean',
    'RETAILPRICE': 'first'
}).reset_index()
```

```python
# getting the count of the SKU's
grouped_data['SKU_COUNT'] = dff['SKU'].transform('count')
```

```python
grouped_data.columns = ['SKU', 'Size', 'Condition', 'Brand', 'Color', 'Silhouette', 'Collab','Average_Price','Retail_Price','SKU_Count']
```

```python
grouped_data.shape
grouped_data.head()
```

```python
# sorting by the SKU count to know whta is the highly sold SKU
grouped_data = grouped_data.sort_values(by=['SKU_Count'], ascending=False)
grouped_data.head()
```

```python
# calculating the Percentage_Difference 
# formula used :- ((Avg_price - Retail_proce) / Retail_price) * 100
grouped_data['Percentage_Difference'] = (grouped_data['Average_Price'] - grouped_data['Retail_Price']) / grouped_data['Retail_Price'] * 100
grouped_data.shape
grouped_data.head()
```

```python
# getting the nulls 
(grouped_data.isnull().sum() / len(grouped_data)) * 100
```

```python
# dropping the nulls to attain the proper percentage diff 
# before dropping - size = 421015
# after dropping - size = 416923
# total loss - 4,092
grouped_data.dropna(inplace=True)
grouped_data.shape
grouped_data.head()
```

```python
grouped_data.sort_values(by=['Percentage_Difference'], ascending=False)
```

```python
# Surprisingly I found there is retail price - 0 
# so i wanted to analyse the data where there is no 0 in the retail price 
# data loss - 11,380
grouped_data_new = grouped_data[grouped_data['Retail_Price'] != 0]
grouped_data_new.shape
grouped_data_new.head()
```

```python
# Add a new column 'Resale_Flag' based on the specified thresholds set on the percentage_difference column
flagged_data=grouped_data_new.copy()

flagged_data['Resale_Flag'] = flagged_data['Percentage_Difference'].apply(lambda x: 'Very_Premium_Resale' if x >= 100 else
                                                           'Premium_Resale' if 75 <= x < 100 else
                                                           'Likely_Premium' if 50 <= x < 75 else
                                                           'Mediumly_Premium' if 25 <= x < 50 else
                                                           'Chance_of_Premium' if 0 < x < 25 else               
                                                           'Not Premium' if -40 <= x < 0 else
                                                           'Not Premium' if x == 0 else
                                                           'Unknown')


flagged_data.shape
flagged_data.head()
```

```python
merged_df = pd.merge(df, flagged_data[['SKU', 'Resale_Flag']], on='SKU', how='left')
merged_df.shape
merged_df.head()
```
